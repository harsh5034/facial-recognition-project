```python
import cv2
import face_recognition
import numpy as np
import os
img=r"WhatsApp Image 2023-11-30 at 9.39.29 AM.jpeg"
image = face_recognition.load_image_file(img)
# Encodings for images in the database
encodings_database = face_recognition.face_encodings(image)
# directory=r"C:\Users\Lenovo\OneDrive\Desktop\New folder\picture"
# for filename in os.listdir(directory):
#     if filename.endswith(".jpg"):
#         file_path = os.path.join(directory, filename)
#         image = face_recognition.load_image_file(file_path)
#         face_locations = face_recognition.face_locations(image)
#         for face_location in face_locations:
#             encodings_database = face_recognition.face_encodings(face_location)
                        
def compare_faces(camera_frame, encodings_database):
    # Find face locations and encodings in the current camera frame
    face_locations = face_recognition.face_locations(camera_frame)
    face_encodings = face_recognition.face_encodings(camera_frame, face_locations)

    # Loop through each face found in the camera frame
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # Compare the face encoding with the database
        matches = face_recognition.compare_faces(encodings_database, face_encoding)

        name = "Unknown"  # Default name if no match is found

        # If a match is found, use the name of the first matching image
        if True in matches:
            first_match_index = matches.index(True)
            name = f"Person {first_match_index + 1}"

        # Draw a rectangle and label on the frame
        cv2.rectangle(camera_frame, (left, top), (right, bottom), (0, 255, 0), 2)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(camera_frame, name, (left + 6, bottom - 6), font, 0.5, (255, 255, 255), 1)

    return camera_frame

if __name__ == "__main__":
    # Open the video camera (0 is usually the default camera)
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)  # Set width
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)  # Set height
    cap.set(cv2.CAP_PROP_FPS, 40)  # Set frame rate


    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()

        # Process the frame and display the result
        processed_frame = compare_faces(frame, encodings_database)
        cv2.imshow("Camera Feed", processed_frame)

        # Break the loop if 'q' key is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the video capture object and close the window
    cap.release()
    cv2.destroyAllWindows()
```
